```python
def get_stats(ids):
  counts = {}
  for pair in zip(ids, ids[1:]):
    counts[pair] = counts.get(pair, 0) + 1
  return counts

stats = get_stats(tokens)
#print(stats)
sorted_stats = sorted((v, k) for k, v in stats.items())
sorted_stats.reverse()
print(sorted_stats)
```
